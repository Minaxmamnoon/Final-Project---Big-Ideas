**Introduction**

In this project, we wanted to find out which neighborhoods are best for driving. To do this, we looked at things that affect driving. One important thing was the number of speed bumps in each neighborhood. Fewer speed bumps usually mean smoother driving with less stopping and starting. So, we counted how many speed bumps are in each neighborhood. This shows which areas have more traffic slow-downs. We looked at other factors too, but we focused on speed bumps because they directly affect how smooth driving feels.

**Description of metrics**

In this part of the project, our metric is to determine which neighborhoods are best based on the number of speed humps. Neighborhoods with fewer speed humps are considered better, while those with more are seen as less desirable.

Datasets: For this analysis, we used the dataset from https://data.wprdc.org/dataset/city-of-pittsburgh-speed-humps, which includes information on all speed humps in Pittsburgh up to 2021.

```python
import pandas as pd
data = pd.read_csv("speed_humps.csv")
data.head(10)
```

```python
neighborhood_counts = data.groupby("neighborhood")['id'].count()
print(neighborhood_counts)
```

```python
neighborhood_counts.plot.bar()
```

**Thought Process**

After loading the dataset on Pittsburgh speed humps, we checked which neighborhoods had the most. Then, we created a bar graph showing how many speed humps each neighborhood has.

**Metric Conclusion**

Squirrel Hill South has the most speed humps, making it the least desirable neighborhood for driving. On the other hand, Central Oakland, Overbrook, and Stanton Heights have the fewest speed humps, making them the best neighborhoods to drive in.
