## Introduction
For my part of the final project, I analyzed the “Non-Traffic Citations” dataset from the Western Pennsylvania Regional Data Center (WPRDC). 

My goal was to determine which Pittsburgh neighborhood is the safest to drive in.

### **Metric: Fewer traffic citations = safer driving**

Traffic citations represent recorded driving violations, so fewer violations suggest safer driving behavior and more violations unsafe driving behavior. 


## Code Explanation 

The raw dataset contains citations along with information about the neighborhood where each citation occurred.

Before analyzing the data, I cleaned it by:
- loaded in data by using pandas 
- Removed non-Pittsburgh neighborhoods
- Sorted neighborhoods from fewest to most citations
- Made a simple bar chart of the top 30 neighborhoods

```python
#### import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("traffic citations .csv")
df = df.dropna(subset=["NEIGHBORHOOD"])

# Remove non-neighborhood rows
df = df[~df["NEIGHBORHOOD"].str.contains("Outside|Unable", case=False)]
df = df[df["NEIGHBORHOOD"] != "Mt. Oliver Boro"]  # remove separate borough

citation_counts = df["NEIGHBORHOOD"].value_counts().reset_index()
citation_counts.columns = ["Neighborhood", "Total_Citations"]

citation_counts_sorted = citation_counts.sort_values("Total_Citations", ascending=True)

citation_counts_sorted.head(10)
```

## Count Citations Per Neighborhood

Next, I counted how many citations appear in each neighborhood.

This helped me compare neighborhoods and identify which ones have the fewest driving violations.


```python
safest = citation_counts_sorted.iloc[0]
print("Safest neighborhood:", safest["Neighborhood"])
print("Total citations:", int(safest["Total_Citations"]))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv("traffic citations .csv")
df = df.dropna(subset=["NEIGHBORHOOD"])
df = df[~df["NEIGHBORHOOD"].str.contains("Outside|Unable", case=False, na=False)]
df = df[df["NEIGHBORHOOD"] != "Mt. Oliver Boro"]

# Count citations
citation_counts = df["NEIGHBORHOOD"].value_counts().reset_index()
citation_counts.columns = ["Neighborhood", "Total_Citations"]

# Sort smallest → largest
citation_counts_sorted = citation_counts.sort_values("Total_Citations", ascending=True)

# Top 30 safest neighborhoods
top30 = citation_counts_sorted.head(30)

plt.figure(figsize=(12,8))
plt.barh(top30["Neighborhood"], top30["Total_Citations"])
plt.xlabel("Total Traffic Citations (Lower = Safer)")
plt.ylabel("Neighborhood")
plt.title("Top 30 Neighborhoods in Pittsburgh (Based on Citations)")
plt.gca().invert_yaxis()
plt.show()

```

## Bar graph on top 30 neighborhoods 
This chart shows the top 30 neighborhoods in Pittsburgh based on citation.  

A lower number means a safer driving environment. Each bar represents the total number of traffic citations below it. The safer the area is to drive in. This visualization makes it easy to compare neighborhoods quickly.
