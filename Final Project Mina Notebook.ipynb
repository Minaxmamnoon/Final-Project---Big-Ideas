**Introduction**

In this project, we wanted to find out which neighborhoods are the easiest to drive in. We looked at different things that can affect driving. One important thing was how many speed bumps each neighborhood has. Fewer speed bumps usually mean smoother driving because you don’t have to stop often. So we counted the speed bumps in each area to see which neighborhoods have more or fewer of them. We checked other things too, but we focused on speed bumps because they directly affect how smooth driving feels.

**Description of metrics**

For this part of the project, our metric is based on how many speed humps are in each neighborhood. We say a neighborhood is “better” if it has fewer speed humps, and “less desirable” if it has more.

Datasets: To do this, we used a dataset from the WPRDC that lists all speed humps in Pittsburgh up to 2021. The dataset can be found at the Pittsburgh Speed Humps page on the WPRDC website. https://data.wprdc.org/dataset/city-of-pittsburgh-speed-humps.

```python
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv("speed_humps.csv")
data.head(10)
```

```python
neighborhood_counts = data.groupby("neighborhood")['id'].count()
print(neighborhood_counts)
```

```python
neighborhood_counts.plot.bar()
```
